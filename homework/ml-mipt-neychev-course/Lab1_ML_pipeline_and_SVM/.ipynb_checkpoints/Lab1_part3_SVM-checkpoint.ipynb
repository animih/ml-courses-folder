## 1.1 Pure models.
First let's try to solve this case with good old Logistic Regression and simple (linear kernel) SVM classifier.

Train LR and SVM classifiers (choose params by hand, no CV or intensive grid search neeeded) and plot their decision regions. Calculate one preffered classification metric.

Describe results in one-two sentences.

_Tip:_ to plot classifiers decisions you colud use either sklearn examples ([this](https://scikit-learn.org/stable/auto_examples/neural_networks/plot_mlp_alpha.html#sphx-glr-auto-examples-neural-networks-plot-mlp-alpha-py) or any other) and mess with matplotlib yourself or great [mlxtend](https://github.com/rasbt/mlxtend) package (see their examples for details)

_Pro Tip:_ wirte function `plot_decisions` taking a dataset and an estimator and plotting the results cause you want to use it several times below

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

lr = LogisticRegression(solver='saga', penalty = 'l2', C = 1.0, max_iter=500) # add some params
svm = SVC(kernel='linear', C = 1.0) # here too

classifiers = [make_pipeline(StandardScaler(), lr),
               make_pipeline(StandardScaler(),svm)]
names = ['Logistic regression', 'SVM']
```

```python
from matplotlib import pyplot as plt
#from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap

# for binary classifcation
def plot_decisions(list_of_classifiers, list_of_names, X, y, show=False):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4)
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = .02
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    assert len(list_of_classifiers) == len(list_of_names), 'len(names) != len(clfs)'
    N = len(list_of_classifiers)
    figure = plt.figure(figsize=(17, 9))
    
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    
    i = 1
    accuracy = []
    if show:
        best_param = []
    
    for name, clf in zip(list_of_names, list_of_classifiers):
        if N != 1:
            ax = plt.subplot(1, N, i)
        else:
            fig, ax = plt.subplots()
        
        clf.fit(X_train, y_train)
        
        if show:
            best_param.append(clf.best_params_)
        
        score = clf.score(X_test, y_test)
        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, x_max] x [y_min, y_max].
        if hasattr(clf, "decision_function"):
            Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
        else:
            Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        
        ax.contourf(xx, yy, np.sign(Z), cmap = cm, alpha = 0.5)
        ax.contourf(xx, yy, Z, cmap = cm, levels=[-1, 0, 1], alpha = 0.5)
        # Plot also the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
                   edgecolors='black', s=25)
        # and testing points
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                   alpha=0.6, edgecolors='black', s=25)

        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        ax.set_title(name)
        ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                size=15, horizontalalignment='right')
        i += 1
        accuracy.append(score)
    if show:
        return accuracy, best_param
    return accuracy
```

```python
score = plot_decisions(classifiers, names, moons_points, moons_labels)
plt.show()
print('Accuracy:', score)
```

In short, we can see that dividing strip for SVM is broader (which it tends to achive). However, both classifiers have bad accuracy score for it is clearly not the case of linearly separable samples

Now use different kernels (`poly`, `rbf`, `sigmoid`) on SVC to get better results. Play `degree` parameter and others.

Write couple of sentences on:

```python
### YOUR CODE HERE
from sklearn.model_selection import GridSearchCV

np.random.seed(1)

svm_poly =[ SVC(kernel='poly', degree=3, C = 1.0),
      SVC(kernel='poly', degree=7, C = 1.0),
      SVC(kernel='poly', degree=9, C = 1.0)]
svm_rest=[SVC(kernel='rbf', C = 1.0),
      SVC(kernel='sigmoid', C = 1.0) ]

param_grid = {
    'svc__C' : np.arange(1, 12, 5),
    'svc__gamma': [1e-3, 1e-4, 1e-5, 'auto']
}

classifiers = []
for clf in svm_poly:
    pipe = make_pipeline(StandardScaler(),clf)
    classifiers.append(GridSearchCV( pipe , param_grid, scoring='accuracy', cv=5))
names = ['Poly_deg=3', 'Poly_deg=7', 'Poly_deg=9']

score, param = plot_decisions(classifiers, names, moons_points, moons_labels, True)
plt.show()
print('Accuracy:', score)
print('Param:', param)

classifiers = []
for clf in svm_rest:
    pipe = make_pipeline(StandardScaler(),clf)
    classifiers.append(GridSearchCV( pipe , param_grid, scoring='accuracy', cv=5))
names = ['RBF', 'Sigmoid']

score, param = plot_decisions(classifiers, names, moons_points, moons_labels, True)
plt.show()
print('Accuracy:', score)
print('Param:', param)

scores = []

X_train, X_test, y_train, y_test = train_test_split(moons_points, moons_labels, test_size=.4)

for i in range(1, 19, 2):
    clf = SVC(kernel='poly', degree=i, C = 1.0)
    pipe = make_pipeline(StandardScaler(),clf)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    scores.append(score)

plt.title('Optimal degree of polynomial kernel')
plt.xlabel('deg')
plt.xticks(np.arange(1, 19, 2))
plt.ylabel('accuracy')
plt.scatter(np.arange(1, 19, 2), scores)
plt.show()
```

The accuracy has increased minorly for polynomial kernel (deg = 3 is optimal). RBF kernel classifier reached accuracy = 0.98, and as we can see on a plot RBF kernel is capable of building comlex division boundaries. For the sigmoid kernel, as expected, we reached approximately the same result as with logistic regression classifier (only the division strip became wider)

Deg = 3 is optimal, because for the bigger degrees our model tend to overlearn

## 1.3 Simpler solution (of a kind)
What is we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Comare SVM's results with this solution (quality, borders type)
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.

```python
from sklearn.preprocessing import PolynomialFeatures
np.random.seed(1)
poly = PolynomialFeatures(degree=5)
pipe = make_pipeline(poly, StandardScaler(), lr)

score = plot_decisions([pipe], ['Polynomial Features deg=5'], moons_points, moons_labels)
plt.show()
print(score)
```

As we can see by generating polynomial features we reached good accuracy and comlex division border. The degree is chosen not too big (<20) to avoid creating too much corellated features

For the same deg=5, SVM with poly kernel gives accuracy 0.84, which is signifactly worse than the result we reached with polynomial features.

```python
### for multiclassification prob

from sklearn.metrics import f1_score

def plot_decisons_mult(clf, name, X, y, show=False):
    
    # labels
    labels = np.unique(y)
    n = labels.shape[0]

    # train test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4)
    
    # meshgrid
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = .02
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    fig, ax = plt.subplots()
    
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    f1 = f1_score(y_test, clf.predict(X_test), average = None)
    
    if show:
        best_param = clf.best_params_
        
    if hasattr(clf, "decision_function"):
        Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    else:
        Z = (clf.predict_proba(np.c_[xx.ravel(), yy.ravel()]))
        
    pred = 2*np.argmax(Z, axis = 1)-3

    pred = pred.reshape(xx.shape)
    ax.contourf(xx, yy, pred, cmap = plt.cm.coolwarm, alpha = 0.5)
    #ax.contourf(xx[mask], yy[mask], pred[mask], cmap = plt.cm.coolwarm, alpha = 0.9)

    # Plot points
    ax.scatter(X[:, 0], X[:, 1], c=2*y-3, cmap = plt.cm.coolwarm,
               edgecolors='black', s=25)

    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(name)
            
    if show:
        return score, f1, best_param
    return score, f1
```

```python
### YOUR CODE HERE

pipe = make_pipeline(StandardScaler(), SVC(kernel='rbf', C = 1.0))

clf = GridSearchCV( pipe , param_grid, scoring='accuracy', cv=5)
name = 'RBF'

score, f1, param = plot_decisons_mult(clf, name, points, labels, True)
plt.show()
print('Accuracy:', score)
print('F1 score:', f1)
print('Param:', param)
```
